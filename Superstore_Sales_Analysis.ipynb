7. Create a Monthly Sales Trend  

```python
monthly_sales=df.groupby(df['Order Date'].dt.month)['Sales'].sum().round(2)
```

```python
plt.figure(figsize=(10,5))
plt.plot(monthly_sales.index, monthly_sales.values, marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.savefig("Monthly_sales_trend.png", dpi=300, bbox_inches='tight')
plt.show()
```

## Phase 6 : Data Visualization

1. Create a Bar Chart showing Sales by Category.

```python
Sales_by_Category=df.groupby('Category')['Sales'].sum().round(2)
plt.figure()
sns.barplot(x=Sales_by_Category.index,y=Sales_by_Category.values,color='royalblue')
plt.title("Categories Wise Sales")
plt.xlabel("Categories")
plt.ylabel('Sales')
plt.savefig("sales_by_categories.png", dpi=300, bbox_inches='tight')
plt.show()
```

2. Create a Bar Chart showing Sales by Region.

```python
Sales_by_region=df.groupby('Region')['Sales'].sum()
plt.figure()
plt.bar(Sales_by_region.index,Sales_by_region.values)
plt.title("Region Wise Sales")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.savefig("sales_by_region.png", dpi=300, bbox_inches='tight')
plt.show()
```

3.Create a Pie Chart showing Sales by Segment.

```python
Sales_by_Segment=df.groupby('Segment')['Sales'].sum()
plt.figure()
plt.pie(Sales_by_Segment.values,labels=Sales_by_Segment.index,autopct='%1.1f%%')
plt.title("Sales by Segment")
plt.savefig("sales_by_segment.png", dpi=300, bbox_inches='tight')
plt.show()
```

4. Create a Line Chart showing the Monthly Sales Trend.

```python
monthly_trend=df.groupby(df['Order Date'].dt.month)['Sales'].sum().round(2)
plt.figure()
plt.plot(monthly_trend,marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

5. Create a Horizontal Bar Chart for the Top 10 Customers by Sales.

```python
top_10_customer_by_sales=df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
plt.barh(top_10_customer_by_sales.index,top_10_customer_by_sales.values)
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.title('Top 10 Customers by Sales')
plt.xlabel('Sales')
plt.ylabel('Customer Name')
plt.savefig("top10_customers.png", dpi=300, bbox_inches='tight')
plt.show()
```

6. Create a Horizontal Bar Chart for the Top 10 Products by Sales.

```python
top_10_product_by_sales=df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))
plt.barh(top_10_product_by_sales.index,top_10_product_by_sales.values)
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.title("Top 10 Product by Sales")
plt.xlabel('Sales')
plt.ylabel('Product Name')
plt.show()
```

7. Create a Histogram for the Sales column.

```python
plt.figure()
plt.hist(df['Sales'],bins=20)
plt.title("Sales Distribution")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

8. Create a Box Plot for the Sales column.

```python
plt.figure(figsize=(8,6))
plt.boxplot(df['Sales'])
plt.title("Sales Box Plot")
plt.ylabel("Sales")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

9. Create a Bar Chart showing the Top 10 Cities by Sales

```python
top_10_city_by_sales=df.groupby('City')['Sales'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(8,6))
plt.barh(top_10_city_by_sales.index,top_10_city_by_sales.values)
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.title("Top 10 Cities by Sales")
plt.xlabel('Sales')
plt.ylabel('Cities Name')
plt.show()
```

10. Create a Line Chart showing the Yearly Sales Trend.

```python
yearly_sales_trend=df.groupby(df['Order Date'].dt.year)['Sales'].sum().round(2)
plt.figure(figsize=(8,6))
plt.plot(yearly_sales_trend,marker='o')
plt.title('Yearly Sales Trend')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.grid(True)
plt.show()
```
